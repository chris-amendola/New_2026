```python
import pandas as pd
import numpy as np
import re
```

```python
path='C:\\Users\\camendol\\Downloads\\'
file='Extract Unpacked v80_Field Descriptions (4).xls'

file_path=path+file
```

```python
xl_file = pd.ExcelFile(file_path)

xl_file.sheet_names  
```

```python
data_dict=xl_file.parse('Sheet1')
```

```python
data_dict.head()
```

```python
data_dict['name']=data_dict['Column']
data_dict['informat']=np.where( data_dict['Native_Format'].isna()\
                               ,('$'+data_dict['Length'].astype(str)+'.')\
                               , '-')

data_dict['format']=np.where( data_dict['Native_Format'].isna()\
                              ,('$'+data_dict['Length'].astype(str)+'.')\
                               ,'-')

data_dict['inc_len']=np.where(data_dict['Native_Format'].isna()\
                              ,(data_dict['Length'].astype(str))\
                               ,'-')

data_dict['label']=data_dict['Description']

#Special Case Vars
##DATE
data_dict['informat']=np.where( data_dict['Native_Format']=='CCYYMMDD'\
                               ,('YYMMDD8.')\
                                ,data_dict['informat'])

data_dict['format']=np.where( data_dict['Native_Format']=='CCYYMMDD'\
                               ,('YYMMDD8.')\
                                ,data_dict['format'])

data_dict['inc_len']=np.where( data_dict['Native_Format']=='CCYYMMDD'\
                               ,('8')\
                                ,data_dict['inc_len'])

##True Numerics
data_dict['informat']=np.where((data_dict['Native_Format'].notnull()) & (data_dict['Native_Format']!='CCYYMMDD')\
                               ,'BEST12.'\
                               ,data_dict['informat'])

data_dict['format']=np.where((data_dict['Native_Format'].notnull()) & (data_dict['Native_Format']!='CCYYMMDD')\
                               ,'BEST12.'\
                               ,data_dict['format'])

data_dict['inc_len']=np.where( (data_dict['Native_Format'].notnull()) & (data_dict['Native_Format']!='CCYYMMDD')\
                               ,(data_dict['Length'].astype(str))\
                               ,data_dict['inc_len'])

data_dict['transform']=' '
data_dict['pad']=' '
```

```python
data_dict.head(50)
```

```python
out_path='C:\\Users\\camendol\\Documents\\'
schema=data_dict[['name','informat','inc_len','format','label','transform','pad']]
outfile=out_path+'test.schema'
schema.to_csv(outfile, sep='|',index=False)
```
