```python
import tabula
import pandas as pd
```

```python
file = "C:\\Users\\camendol\\Documents\\Unpacked Extract - v8.0E.pdf"
#df = tabula.read_pdf(file ,pages='all',guess=True ,multiple_tables=True)
df=pd.DataFrame([])
df_combine=pd.DataFrame([])
for pageiter in range(5,50):
            #print(pageiter)
            df = tabula.read_pdf(file,pages=pageiter+1, guess=True,pandas_options={'header':None})
            print(df.columns)
            #If you want to change the table by editing the columns you can do that here.
            df_trans=df.loc[(df[0].notnull())&(df[0]!='ROW')]
            #df_trans['row']= df[0].astype(int)
            df_combine=pd.concat([df_trans,df_combine],) #again you can choose between merge or concat as per your need

excel_book='H:\\stuff.xlsx'

with pd.ExcelWriter(excel_book) as writer:
    pd.DataFrame(df_combine)\
      .to_excel(writer, sheet_name='Raw')
```
