```python
import pandas as pd
import numpy as np
import re
import datetime as dt
from pivottablejs import pivot_ui
import pivottablejs
import qgrid
import matplotlib.pyplot as plt
```

```python
path='C:\\Users\\camendol\\Desktop\\'
file='PDQs.xlsx'

file_path=path+file
```

```python
#Open connection to EXCEL
xl_file = pd.ExcelFile(file_path)
xl_file.sheet_names 
```

```python
#Read a tab
dql_data=xl_file.parse('PDQs')
```

```python
dql_data.info(verbose=True)
```

```python
label_cols=dql_data.filter(regex='^Label',axis=1).columns
dql_data['all_labels']=dql_data.loc[:,label_cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
dql_data['PDQ']=np.where( dql_data.all_labels.str.contains('pdq|PDQ') ,1 ,0) 
dql_data['CDQ']=np.where( dql_data.all_labels.str.contains('cdq|CDQ') ,1 ,0) 
dql_data['Implementation']=np.where( dql_data.all_labels.str.contains('Implementation|implementation') ,1 ,0)
dql_data['Production']=np.where( dql_data.all_labels.str.contains('Production|production') ,1 ,0)
dql_data['Delay']=np.where( dql_data.all_labels.str.contains('Delay|delay') ,1 ,0)
```

```python
selected=dql_data[[ 'Project key'\
                   ,'Summary'\
                   ,'Custom field (Group ID)'\
                   ,'Issue Type'\
                   ,'Issue key'\
                   ,'Project type'\
                   ,'Status'\
                   ,'Fix Version/s'\
                   ,'Created'\
                   ,'Updated'\
                   ,'Last Viewed'\
                   ,'Resolved'\
                   ,'Custom field (Existing Client Migration or New Client Implementation?)'\
                   ,'PDQ'\
                   ,'CDQ'\
                   ,'Implementation'\
                   ,'Production'\
                   ,'Delay'
                   
]]
#Project Key Project name Project type Summary Status
selected.head()
```

```python
pivot_ui(selected,rendererName='Heatmap')
```

```python
qgrid.show_grid(selected[['Delay','Production','Implementation','Project key','Summary','Issue key']],show_toolbar=True)
```
